## Las Variables
1. CP --> Costo de Productos
2. CD --> Cantidad de dinero
3. Cambio
## Formulas
$Cambio = CD - CP$
## Diagrama de flujo lucidchart
<img src="https://raw.githubusercontent.com/DanielGallegos128/Calcular-cambio-if-else/refs/heads/main/DF%20if%20else%20Cambio%202.png" alt="Texto alternativo" width="350" height="500">

```python
# Para cuando es verdadero (si)
# Calculando el cambio usando if else
# Datos de entrada CP , CD
# Mostrar Cambio si es falso
# Si es verdadero mostrar "No alcanza"
CP = float(input("Ingrese el costo de los productos: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Prograna realizado por Daniel Gallegos NL = 20")
```

```python
# Para cuando es falso (no)
# Calculando el cambio usando if else
# Datos de entrada CP , CD
# Mostrar Cambio si es falso
# Si es verdadero mostrar "No alcanza"
CP = float(input("Ingrese el costo de los productos: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Prograna realizado por Daniel Gallegos NL = 20")
```
